## Computando a mediana



**Problema**: encontrar a mediana de uma lista ordenada.

**Entrada**: uma lista contendo um ou mais valores numéricos ordenados em ordem crescente.

**Saída**: um número representando a mediana dos valores da lista de entrada.

**Sugestão**: se informe sobre como calcular a mediana. Existe uma consideração importante que deve ser levada em conta dependendo do número de itens da lista.

**Alguns exemplos de entradas e retornos**:


| Entrada <code>lista_ordenada</code>|Retorno|
|---|---|
| [1, 2, 3, 4, 5]|3|
| [1, 2, 3, 4]|2.5|
| [-100, 1, 2]|1|
| [10]|10|
| [0, 0, 0, 0, 5, 6, 7, 8]|2.5|

**Algoritmo**:



```python
def encontra_mediana(lista_ordenada):

    #cálculo da mediana para listas com tamanho par
    if len(lista_ordenada) % 2 == 0:
        mediana = (lista_ordenada[(len(lista_ordenada)//2)] + lista_ordenada[(len(lista_ordenada)//2) - 1])/2
    #cálculo da mediana para listas com tamanho ímpar
    else:
        mediana = lista_ordenada[len(lista_ordenada)//2]
    
    return mediana
```

```python
#TESTE MEDIANA
print(encontra_mediana([1, 2, 3, 4, 5]))
print(encontra_mediana([1, 2, 3, 4]))
print(encontra_mediana([-100, 1, 2]))
print(encontra_mediana([10]))
print(encontra_mediana([0, 0, 0, 0, 5, 6, 7, 8]))
```

## Frequência dos valores



**Problema**: encontrar a frequência de valores de uma lista.

**Entrada**: uma lista numérica de tamanho $n \geq 0$.

**Saída**: um dicionário onde as chaves representam itens da lista de entrada (sem repetição) e os valores representam a quantidade que cada item aparece na lista de entrada (a quantidade nunca pode ser igual a zero no dicionário). 

**Alguns exemplos de entradas e retornos**:


| Entrada <code>lista</code>|Retorno|
|---|---|
| [1, 2, 3]|{1: 1, 2: 1, 3: 1}|
| [2, 2, 3]|{2: 2, 3: 1}|
| [7]|{7: 1}|
| [0, 0, -1, 0, 10, 0, -1]|{0: 4, -1: 2, 10: 1}|
| []|{}|

**Algoritmo**:



```python
def computa_frequencia(lista):
    
    #atribuindo um dicionário vazio como variável local
    frequencia = {}
    #criando loop for para verificar a frequencia dos numeros na lista e adicionar ao dicionario
    for numeros in lista:
        if numeros in frequencia:
            frequencia[numeros] += 1
        else:
            frequencia[numeros] = 1
    
    return frequencia
```

```python
#TESTE FREQUENCIA
print(computa_frequencia([1, 2, 3]))
print(computa_frequencia([2, 2, 3]))
print(computa_frequencia([7]))
print(computa_frequencia([0, 0, -1, 0, 10, 0, -1]))
print(computa_frequencia([]))
```

## Soma específica



**Problema**: dada uma lista que contém apenas números, identificar os índices de dois valores que, quando somados, resultam em um certo valor dado.

**Entradas**: lista numérica contendo apenas números inteiros e um número inteiro $n$.

**Saída**: uma tupla contendo dois números inteiros em ordem crescente. Estes números são diferentes e representam índices da lista de entrada. Quando somamos os elementos da lista de entrada referentes a estes índices, obtemos o valor de $n$. Retornar `False` caso não seja possível construir a soma desejada com os números da lista. Pode retornar *qualquer* resposta que satisfaça as condições para o caso de existir mais de uma resposta (ver como exemplo as últimas duas linhas da tabela abaixo). No entanto, observe que o enunciado pede apenas uma resposta como retorno, logo não é para listar *todas* as respostas possíveis.

**Lembrete**: o Python considera que o primeiro elemento de uma lista tem índice zero.

**Alguns exemplos de entradas e retornos**:


| Entrada <code>lista</code>|Entrada <code>n</code>|Retorno|
|---|---|---|
| [1, 2, 3]|5|(1, 2)|
| [1, 2, 3]|3|(0, 1)|
| [1, 2, 3]|6|False|
| [0, 0, -1, 0, 10, 0, -1]|9|(2, 4)|
| [0, 0, -1, 0, 10, 0, -1]|9|(4, 6)|

**Algoritmo**:



```python
def soma_especifica(lista, n):

    #abrindo loop for 2 vezes com enumerate para verificar os indices que contém valores que somados resultam em n
    for indice, numero in enumerate(lista):
        for indice2, numero2 in enumerate(lista):
            if indice != indice2 and numero + numero2 == n:
                soma = (indice, indice2)
                return soma
            
    return False
```

```python
#TESTE SOMA ESPECÍFICA
print(soma_especifica([1, 2, 3], 5))
print(soma_especifica([1, 2, 3], 3))
print(soma_especifica([1, 2, 3], 6))
print(soma_especifica([0, 0, -1, 0, 10, 0, -1], 9))
print(soma_especifica([0, 0, -1, 0, 10, 0, -1], 9))
```

## Intersecção de duas listas



**Problema**: encontrar itens em comum entre duas listas.

**Entrada**: duas listas numéricas.

**Saída**: uma lista representando a intersecção das listas de entrada. Caso hajam elementos repetidos presentes *nas duas listas*, eles devem estar repetidos também no retorno (ver exemplos na tabela abaixo).

**Alguns exemplos de entradas e retornos**:


| Entrada <code>lista1</code>|Entrada <code>lista2</code>|Retorno|
|---|---|---|
| [1, 2, 3]|[5, 3, 6, 7]|[3]|
| [10, 50, 100]|[-2, 7, 4]|[]|
| [22, -5, 90, 37, 2, 2]|[2, 49, 90, -5]|[2, 90, -5]|
| [22, -5, 90, 37, 2, 2]|[2, 49, 2, 90, -5]|[2, 2, 90, -5]|
| [22, -5, 90, 37, 2, 2, 2]|[2, 49, 2, 90, -5, -5]|[2, 2, 90, -5]|

**Algoritmo**:



```python
def interseccao(lista1, lista2):

    #criando lista vazia como variável local; representará a intersecção entre as listas
    intersecção = []
    #abrindo loop for para percorrer os elementos de cada lista e, depois, comparar os valores e adicionar na variavel intersecção aquels que forem iguais e remover esse elemento da outra lista para evitar repetição
    for n1 in lista2:
        for n2 in lista1:
            if n1 == n2:
                intersecção.append(n1)
                lista1.remove(n2)
                break
            

    return intersecção
```

```python
#TESTE INTERSECÇÃO
print(interseccao([1, 2, 3], [5, 3, 6, 7]))
print(interseccao([10, 50, 100], [-2, 7, 4]))
print(interseccao([22, -5, 90, 37, 2, 2], [2, 49, 90, -5]))
print(interseccao([22, -5, 90, 37, 2, 2], [2, 49, 2, 90, -5]))
print(interseccao([22, -5, 90, 37, 2, 2, 2], [2, 49, 2, 90, -5, -5]))
```
